## Matrix Operation

```python
import numpy as np
```

### 1. Matrix Creation

```python
all_zero = np.zeros((3,2))
cloned_zero = all_zero.copy()
all_one = np.ones((2,3))
increasing = np.arange(10)
diag = np.eye(5)

print(f"All Zero:\n{all_zero}\n\n All One:\n{all_one}\n\n Increasing:\n{increasing}\n\n Diagnal:\n{diag}\n\n")
```

### 2. Numerical Operations on Matrices

#### 2.1 Basic Numerical Operations

```python
m1 = np.array([
    [1,2],
    [3,1],
])

m2 = np.array([
    [2,4],
    [2,1],
])

sum = m1 + m2
dif = m1 - m2
mul = m1 * m2
div = m1 / m2

print(f"sum:\n{sum}\n\ndif:\n{dif}\n\nmul:\n{mul}\n\ndiv:\n{div}\n\n")
```

#### 2.2 Power and Square Root

```python
pow = m1 ** 2
sqr = np.sqrt(m1)

print(f"pow:\n{pow}\n\nsqr:\n{sqr}\n\n")
```

#### 2.3 Matrix Operations

##### 2.3.1 Matrix Multiplication

```python
mat_mul = m1 @ m2
mat_mul_np = np.matmul(m1, m2)
print(f"mattrix multiplication:\nUsing'@':\n{mat_mul}\n\nUsing numpy:\n{mat_mul_np}\n\n")
```

##### 2.3.2 Transpose

```python
before = np.random.randn(4, 5)
after = before.transpose(1,0)
print(f"Transpose:\nBefore:\n{before}\n\nAfter:\n{after}")
```

##### 2.3.3 Reshape

```python
before = np.arange(20)
after = before.reshape(4,5)
print(f"Reshape:\nBefore:\n{before}\n\nAfter:\n{after}\n\n")
```

##### 2.3.4 Indexing

```python
before = np.arange(20)
np.random.shuffle(before)
before = before.reshape(4,5)
print(f"Indexing:\nOriginal:\n{before}\n\nArgmax:\n{np.argmax(before)}\n\nAxis=0:\n{np.argmax(before, axis=0)}\n\nAxis=1:\n{np.argmax(before, axis=1)}\n\n")
```

##### 2.3.5 Slicing

&emsp; ":" means "from-to" regarding one dimension. 
- `2:3` means "from 2 to 3".
- `:3`  means "from start to 3"
- `:-3` means "from start to the last 3"
- `-3:` means "from end the 3rd, i.e., the last three"
- `:`   means "from start to end, i.e., all"

```python
ori = np.arange(20).reshape(4,5)
first_two_rows = ori[:2]
last_three_cols = ori[:,-3:]
print(f"Slicing\nori:\n{ori}\n\nfirst_two_rows:\n{first_two_rows}\n\nlast_three_cols:\n{last_three_cols}\n\n")
```

```python
ori[:,-3:]
```

```python
m1 = ori[1]
m2 = ori[1:2]
print(f"ori[1] ({m1.shape}):\n{m1}\n\nori[1:2] ({m2.shape}):\n{m2}\n\n")
```

##### 2.3.6 Value Updating

```python
ori = np.arange(20).reshape(4,5)
print(f"Ori:\n{ori}\n\n")

ori[2] += 100
print(f"Add 100 at the 3rd row:\n{ori}\n\n")

ori[:, :2] *= -1
print(f"Reverse first two columns:\n{ori}\n\n")
```

##### 2.3.7 Matrix Concatenation

&emsp; Completely concatenating the rows(axis=0) or columns(axis=1) into a single uniformed matrix.

```python
ori = np.arange(20).reshape(4,5)
copy = np.ones_like(ori) # A matrix filled with ones with the shape of ori

# Concatenate
ori_copy_concat_0 = np.concatenate([ori, copy], axis=0)
ori_copy_concat_1 = np.concatenate([ori, copy], axis=1)

print(f"Concatenate:\nAxis=0:\n{ori_copy_concat_0}\n\nAxis=1:\n{ori_copy_concat_1}\n\n")
```

##### 2.3.8 Matrix Stacking

&emsp; Take the two matrices as discrete elements, and put them into an array.

```python
# Stack
ori_copy_stack_0 = np.stack([ori, copy], axis=0)
ori_copy_stack_1 = np.stack([ori, copy], axis=1)
print(f"Stack:\nAxis=0:\n{ori_copy_stack_0}\n\nAxis=1:\n{ori_copy_stack_1}\n\n")
```

##### 2.3.9 Auto Shape Broadcasting

&emsp; Automatically adjust the shape of scalars/matrix.

```python
ori = np.arange(10)
new = ori + 10
print(f"Ori:\n{ori}\n\nNew:\n{new}\n\n")
```

&emsp; However, improper usage of of auto shape broadcasting will cause undesired results.

```python
m1 = np.arange(5).reshape(1,5)
m2 = np.arange(5).reshape(5,1)
sum = m1 + m2
print(f"m1:\n{m1}\n\nm2:\n{m2}\n\nsum:\n{sum}\n\n")
```
