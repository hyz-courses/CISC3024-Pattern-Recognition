```python
import numpy as np
```

# Matrix Basics

```python
all_zero = np.zeros((3,2))
cloned_zero = all_zero.copy()
print(f"all zero ({all_zero.shape}):\n{all_zero}\n cloned zero ({cloned_zero.shape}):\n{cloned_zero}")
```

```python
all_one = np.ones((2,3))
print(f"all one ({all_one.shape}):\n{all_one}")
```

```python
increasing = np.arange(10)
print(f"increasing ({increasing.shape}):\n{increasing}")
```

```python
diag = np.eye(5)
print(f"diagnal ({diag.shape}):\n{diag}")
```

# Numerical Operations

```python
a = np.array([
    [1,2],
    [3,1]
])

b= np.array([
    [1,2],
    [3,1]
])

sum = a + b
dif = a - b
mul = a * b
div = a / b
pow = a ** 2
sqr = np.sqrt(a)

print(f"a:\n{a}\nb:\n{b}\nsum:\n{sum}\ndif:\n{dif}\nmul:\n{mul}\ndiv:\n{div}\npow:\n{pow}\nsqr:\n{sqr}\n")
```

```python
i = np.exp(a)
print(f"e's power:\n{i}")
```

# Matrix Operations

## Matrix Multiplication

```python
mat_mul_1 = a @ b
mat_mul_2 = np.matmul(a,b)
print(f"Original Matrices:\n{a}\n{b}\nMatrix Multiplication\n Using '@':\n{mat_mul_1}\n Using np:\n{mat_mul_2}\n")
```

## Matrix Transpose

```python
# Randomize a 5-d matrix
rand = np.random.randn(2, 3, 4, 5, 6)
rand_transposed = rand.transpose(3, 0, 1, 4, 2)

print(f"Random Matrix:\n{rand}\nTransposed Matrix:\n{rand_transposed}")
```

## Matrix Reshape

```python
# Reshaping
before = np.arange(20)
after = before.reshape(4,5)
print(f"Before:\n{before}\nAfter:\n{after}\n")
```

# Matrix Index

```python
m0 = np.arange(5)
m1 = np.ones_like(m0)
m2 = np.concatenate([m0, m1], axis=0)
m3 = np.stack([m0, m1], axis=0)
m4 = np.stack([m0, m1, m0], axis=1)

print(f"m0:\n{m0}\nm1:\n{m1}\nm2:\n{m2}\nm3:\n{m3}\nm4:\n{m4}\n")
```

## Shape Broadcasting

```python
# Success
before = np.arange(10)
after = before + 10
print(f"Before:\n{before}\nAfter:\n{after}\n")
```

```python
# Unexpected
m0 = np.arange(5).reshape(1,5)
m1 = np.arange(5).reshape(5,1)
sum = m0 + m1
print(f"\nm0:\n{m0}\nm1:\n{m1}\nsum:\n{sum}\n")
```
